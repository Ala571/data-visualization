```python
data.update((data['Age']).fillna(value=(data['Age']).mean(), inplace = True))
```

```python
data1=data.dropna(axis=1,how='any',thresh=400)
data1.info()
```

```python
data2=data1.dropna(axis=0)
data2
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()

import matplotlib.pyplot as plt
import seaborn as sns
data3=data2.drop(['PassengerId', 'Survived', 'Pclass', 'Name','SibSp','Parch','Ticket', 'Fare', 'Embarked'],axis=1)
data3['Sex']=encoder.fit_transform(data3['Sex'])
Scaler= StandardScaler()
data3['Age']=Scaler.fit_transform(data3[['Age']].values)
data3['Sex']=Scaler.fit_transform(data3[['Sex']].values)
sns.boxplot(data=data3)
```

```python
def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )

plot_correlation_map(data2)
```
